## 저장

```python
from ultralytics import YOLO
import cv2

# YOLO 모델 로드
model = YOLO("경로/best_growth_model.pt")

# 비디오 파일 열기
cap = cv2.VideoCapture("비디오경로")

# 비디오의 프레임 속도와 크기 가져오기
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# 저장할 비디오 파일 설정
output_file = "output_video.mp4"
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter(output_file, fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()

    if ret:
        # YOLO 모델로 프레임 처리
        results = model(frame, conf=0.25, iou=0.7)
        output_frame = results[0].plot()

        # 처리된 프레임을 저장할 비디오 파일에 쓰기
        out.write(output_frame)

        # 결과 프레임 표시
        cv2.imshow("Results", output_frame)
        
        # 'q' 키를 누르면 프로그램 종료 & 누른 시점까지 녹화, 저장
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

cv2.destroyAllWindows()
cap.release()
out.release()
```
