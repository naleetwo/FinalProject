## YOLO 기준 컬럼 추가

### cx, cy, w, h

- cx = (xtl+xbr)/2
- cy = (ytl+ybr)/2
- w = xbr-xtl
- h = ybr-ytl

```python
yolo_df["cx"] = (yolo_df["xtl"] + yolo_df["xbr"])/2
yolo_df["cy"] = (yolo_df["ytl"] + yolo_df["ybr"])/2
yolo_df["w"] = yolo_df["xbr"] - yolo_df["xtl"]
yolo_df["h"] = yolo_df["ybr"] - yolo_df["ytl"]
```

```python
yolo_df
```
