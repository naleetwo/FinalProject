```python
# tom3 환경 데이터 가져오기 
# 파일 경로
file_path = "C:/seokwonna/pythonwork/AI/FinalProject/097.지능형 스마트팜 통합 데이터(토마토)/01.데이터/1.Training/원천데이터/test/TS_Timeseries/tom3/01_(함수율저울,열화상센서,작물환경센서)/환경"

# 파일 목록 가져오기
total_file_list = glob.glob(os.path.join(file_path, '*.csv'))

# 데이터프레임 초기화
merged_df = pd.DataFrame()

# 각 파일을 읽어와서 데이터프레임에 추가
for file in total_file_list:
    df = pd.read_csv(file, sep=";", header=2)
    merged_df = pd.concat([merged_df, df], axis=0)

# 결과 출력
merged_df
```
