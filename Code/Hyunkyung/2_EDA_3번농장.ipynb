# json & image 매칭

```python
TRAIN_LABELING_DIR = os.path.join(TRAIN_DATA_DIR, '라벨링데이터')
```

```python
TL_LIST = os.listdir(TRAIN_LABELING_DIR)
TL_LIST.sort()
TL_LIST
```

```python
TRAIN_SOURCEDATA_DIR = os.path.join(TRAIN_DATA_DIR, '원천데이터')
```

```python
TS_LIST = os.listdir(TRAIN_SOURCEDATA_DIR)
TS_LIST = TS_LIST[1:]
TS_LIST.sort()
TS_LIST
```

```python
for i in range(len(TS_LIST)):
    with zipfile.ZipFile(os.path.join(TRAIN_LABELING_DIR, TL_LIST[i]), 'r') as zip_file:
        json_file_list = zip_file.namelist()

    json_file_list = [item.replace('.json', '') for item in json_file_list]
    print(len(json_file_list))

    with zipfile.ZipFile(os.path.join(TRAIN_SOURCEDATA_DIR, TS_LIST[i]), 'r') as zip_file:
        image_file_list = zip_file.namelist()

    image_file_list = [item.replace('.png', '') for item in image_file_list]
    print(len(image_file_list))

    for element in json_file_list:
        if element not in image_file_list:
            print(False)
    print(True)
    print(i, '\n===========================')
```
