```python
total = pd.DataFrame()

for i in range(len(TL_LIST)):
    
    file_list = os.listdir(os.path.join(TRAIN_LABELING_DIR, TL_LIST[i]))
    file_list
    print(TL_LIST[i], len(file_list))

    df = pd.DataFrame(file_list, columns=['file_name'])

    df['farmid'] = df.file_name.apply(lambda x:x.split('_')[1])
    df['week'] = df.file_name.apply(lambda x:x.split('_')[2])
    df['crop_id'] = df.file_name.apply(lambda x:x.split('_')[3])
    df['part'] = df.file_name.apply(lambda x:x.split('_')[4])
    df['flowercluster'] = df.file_name.apply(lambda x:x.split('_')[5])
    df['date'] = df.file_name.apply(lambda x:x.split('_')[6])
    df

    new_columns=['file_name', 'farmid', 'crop_id', 'part', 'flowercluster', 'week', 'date']

    df = df.reindex(columns=new_columns)
    
    total = pd.concat([total, df], axis=0)
```

```python
sorted_total = total.sort_values(by=['farmid', 'crop_id', 'date', 'week'])
sorted_total
```
